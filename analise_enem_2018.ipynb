```python
import json
br_estados = 'br_states.json'
geo_json_data = json.load(open(br_estados))
```

```python
import folium
```

```python
mapa_inscritos_brasil = folium.Map([-15.77972, -47.92972],  tiles='cartodbpositron', zoom_start=4)
```

```python
dt_2018['CO_MUNICIPIO_PROVA_INICIAL'] = dt_2018['CO_MUNICIPIO_PROVA'].astype(str).str[:2]
dt_2018
```

```python
dt_2018_PROVA = pd.DataFrame(data = dt_2018['CO_MUNICIPIO_PROVA_INICIAL'].value_counts().sort_values(ascending=False)).reset_index()
dt_2018_PROVA.columns = ['UF', 'Inscritos']
dt_2018_PROVA['Sigla'] = ['SP', 'MG', 'BA', 'RJ', 'CE', 'PE', 'PA', 'RS', 'PR', 'MA', 'GO', 'PB', 'SC', 'RN', 'AM', 'PI', 'ES', 'DF', 'MT','AL','SE','MS', 'RO','TO', 'AP', 'AC', 'RR']
dt_2018_PROVA
```

```python
folium.Choropleth(geo_data = geo_json_data,
                 data = dt_2018_PROVA,
                 columns = ["Sigla", "Inscritos"],
                 key_on = "feature.id",
                 fill_opacity = 0.9,
                 line_opacity = 0.5,
                 legend_name = "Inscritos por Região -  ENEM 2018",
                 nan_fill_color = "white",
                 name = "Dados").add_to(mapa_inscritos_brasil)
estilo = lambda x: {'fillcolor': 'white',
                    'color': 'black',
                     'fillOpacity': 0.001,
                      'weight': 0.001 }

estilo_destaque = lambda x: {'fillcolor': 'blue',
                    'color': 'blue',
                     'fillOpacity': 0.5,
                      'weight': 1 }

highlight = folium.features.GeoJson(data=geo_json_data, style_function=estilo, highlight_function = estilo_destaque )
folium.features.GeoJsonTooltip(fields = ["name"],
                              aliases = ["Sigla"],
                               labels = False,
                              style = ("background-color: white; color: black; font-family: arial; font-size: 16px; padding: 10px;")).add_to(highlight)

mapa_inscritos_brasil.add_child(highlight)

mapa_inscritos_brasil
```
